```python
import requests
import json
import pandas as pd

meatbox_url = "https://www.themeatbox.co.nz/collections/all-meat/products.json?limit=300"

meatbox_result = requests.get(meatbox_url)
meatbox_data = meatbox_result.json()

meatbox_products = []

for item in meatbox_data["products"]:
    title = item["title"]
    handle = item["handle"]
    # print(title, handle)
    for variant in item["variants"]:
        price = variant["price"]
        available = variant["available"]
        # print(price, available)

    product = {
        "title" : title,
        "handle": handle,
        "price": price,
        "available": available
    }
    meatbox_products.append(product)

df = pd.DataFrame(meatbox_products)
df.to_csv("Meatbox prices.csv")
print("All done!")
```
